```python
import requests

# API URL
url = "https://developer.nps.gov/api/v1/parks"
api_key = "pmawI9sL3Sj6Qf49wTWrrRz3NKsl6Fe415X8bq7x"

# Parameters
params = {
    "api_key": api_key,
    "fields": "name,states,visitor_count"
}

# Send GET request to the API
response = requests.get(url, params=params)

# Check if the request was successful
if response.status_code == 200:
    parks_data = response.json()
    parks = parks_data["data"]

    # Print the relation between Number of Visitors per Year and State
    for park in parks:
        name = park["name"]
        states = park["states"]
        visitor_count = park["visitor_count"] if "visitor_count" in park else "N/A"

        print("Park Name:", name)
        print("State(s):", states)
        print("Number of Visitors per Year:", visitor_count)
        print("--------------------")
else:
    print("Failed to retrieve park data. Status code:", response.status_code)
```
